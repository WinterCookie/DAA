```python
def getsum(mylist):
  sum = 0
  for item in mylist:
    sum = sum + item
  return sum
getsum([1,2,3,4])
```

```python
def getbagi(mylist):
  bagi = 0
  for item in mylist:
    bagi = bagi + item
  return bagi
getbagi([1,2,3,4])/4
```

```python
def getkali(mylist):
  kali = 1
  for item in mylist:
    kali = kali * item
  return kali
getkali([1,2,3,4])
```

```python
def getsum(mylist):
    sum = 0
    for row in mylist:
        for item in row:
            sum += item
    return sum
getsum([[1,2,5],[3,4,7]])
```

```python
def getkurang(mylist):
    kurang = 0
    for row in mylist:
        for item in row:
            kurang -= item
    return kurang
getkurang([[1,2,5],[3,4,7]])
```
